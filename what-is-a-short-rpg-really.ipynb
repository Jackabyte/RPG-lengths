```python
hltb = pd.read_csv("/kaggle/input/howlongtobeat-games-completion-times/games.csv")

hltb.head()
```

Fill type coloumn to replace NaN with Game

```python
values = {"type": 'Game'}

hltb = hltb.fillna(value=values)
hltb
```

Filter so we only have RPGs (Roll Playing Games)

```python
RPGS = hltb[hltb['genres'].str.contains("Role-Playing|role-playing|Role-playing|role-Playing")==True]
RPGS.shape
```

Filter so we only have Games (DLC, Expansions and Mods are removed)

```python
RPGS = RPGS[RPGS['type'] == 'Game']
RPGS.shape

#RPGS = hltb[hltb['genres'].str.contains("Turn-based|Turn-Based")==True]
#RPGS.shape
```

```python
#hltb = hltb[hltb['title'], hltb['main_story'], hltb['main_plus_extras'],hltb['genres']]

RPGS = RPGS.loc[:, ['title', 'main_story', 'main_plus_extras', 'genres']]
RPGS = RPGS.set_index('title')
RPGS
```
